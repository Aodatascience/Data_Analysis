# Genere wise Preference of Movies 

```python
import pandas as pd
import numpy as np
import os
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
os.chdir(r"C:\Users\Alok\Desktop\MBA")
```

```python
df=pd.read_csv('Movie.csv',header=None)
df.head()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,4) if str(df.values[i,j])!='0'])
```

```python
transactions
```

```python
from mlxtend.preprocessing import TransactionEncoder

x = TransactionEncoder()
y = x.fit_transform(transactions)
df = pd.DataFrame(y, columns = x.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True,verbose=1)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='lift')  
rules.sort_values('lift', ascending=False)
```

```python
Output=round(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']],3)
```

```python
Output
```

```python
targeted_rules=Output[Output["consequents"] == {'Action'}]
print(targeted_rules)
```

```python
filtered_rules = targeted_rules[(targeted_rules['confidence'] > 0.5) &
                                (targeted_rules['lift'] > 1.00)]
filtered_rules
```
